# Nodes choosing

```python
# ── 10 個節點 + CHT (台達電沒有ADR故移除）──────────────────────────────────────
ADR_TICKERS = [
    "TSM",    # 台積電
    "UMC",    # 聯電
    "ASX",    # 日月光
    "AUOTY",  # 友達
    "HIMX",   # 奇景
    "SIMO",   # 慧榮
    "IMOS",   # 南茂
    "CHT",    # 中華電信（L1 保留）
    "PWR",  # 廣達
    #"DTLYY",  # 台達電
    "ASUUY",  # 華碩
    "AMADY",  # 研華
]
 
TW_TICKERS = [
    "2330.TW",  # 台積電
    "2303.TW",  # 聯電
    "3711.TW",  # 日月光
    "2409.TW",  # 友達
    "3596.TW",  # 奇景
    "6415.TW",  # 慧榮
    "8150.TW",  # 南茂
    "2412.TW",  # 中華電信（L1 保留）
    "2382.TW",  # 廣達
    #"2308.TW",  # 台達電
    "2357.TW",  # 華碩
    "2395.TW",  # 研華
]
 
SENTIMENT_TICKERS = {
    "VIX":    "^VIX",
    "SP500":  "^GSPC",
    "NASDAQ": "^IXIC",
    "TAIEX":  "^TWII",
    "USDTWD": "TWD=X",
}
```

```python
# ── 下載函數 ─────────────────────────────────────────────
def download_one(ticker: str, start: str, end: str) -> pd.DataFrame:
    """下載單一標的，回傳原始 OHLCV DataFrame"""
    df = yf.download(
        ticker,
        start=start,
        end=end,
        auto_adjust=True,   # 還原權息（前復權）
        progress=False,
    )
    # 若為 MultiIndex，攤平為單層欄位
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    return df
 
 
def save_raw(df: pd.DataFrame, path: str, ticker: str):
    """儲存並印出狀態"""
    if df.empty:
        print(f"  ❌  {ticker:<10} 無資料")
        return
    df.to_csv(path)
    print(f"  ✅  {ticker:<10} {len(df):4d} 筆  "
          f"{df.index[0].date()} ～ {df.index[-1].date()}  "
          f"→ {path}")
```

# Data download

```python
print("=" * 60)
print(f"原始資料下載  {START} ～ {END}")
print("=" * 60)
 
# ADR
print("\n[1/3] ADR（美股）")
for ticker in ADR_TICKERS:
    df = download_one(ticker, START, END)
    save_raw(df, f"/Users/lichengjun/Desktop/Multiplex_gnn_stock_prediction/data/raw/adr/{ticker}.csv", ticker)
    time.sleep(0.4)   # 避免被限速
 
# 台股
print("\n[2/3] 台股（TWSE / TPEx）")
for ticker in TW_TICKERS:
    code = ticker.replace(".TW", "").replace(".TWO", "")
    df = download_one(ticker, START, END)
    save_raw(df, f"/Users/lichengjun/Desktop/Multiplex_gnn_stock_prediction/data/raw/tw/{code}.csv", ticker)
    time.sleep(0.4)
 
# 市場情緒指標
print("\n[3/3] 市場情緒指標")
for name, ticker in SENTIMENT_TICKERS.items():
    df = download_one(ticker, START, END)
    save_raw(df, f"/Users/lichengjun/Desktop/Multiplex_gnn_stock_prediction/data/raw/sentiment/{name}.csv", ticker)
    time.sleep(0.4)
```
